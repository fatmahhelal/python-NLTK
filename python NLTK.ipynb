### Question 3: Omit a given list of stop words ('again', 'once' and 'from')  from the total stopwords list of English language.

```python
# First, bring all English StopWord, and set it in a variable 
eng=SW.words('english')
print(f' The Total English StopwWords is: {len(eng)} ')
# print(eng)
```

```python
# set a variable for the list of words we want to remove
DropList=['again', 'once', 'from']
```

```python
# go iterating over our list to check it one by one
# if the word is available in StopWords list will delete them
for i in DropList:
    if i in eng:
       eng.remove(i)
    print(f'{i} is omitted from StopWords List')
# check out the StopWords list after deleted if it change.
print(f'Total English StopWords after Omit a given list is: {len(eng)} \n\n')
print(eng)
```

### Question 4: Print the definition and examples of any one English language word using WordNet corpus.

```python
# import wordnet from nltk.corpus, 
# set a shortcut as dic for easy use later 
from nltk.corpus import wordnet as dic
```

```python
# ask user to enter one English word to prit its definition and examples
inputWord=input('Please enter one English word :\n')
syns = dic.synsets(inputWord)
print(f"Defination of {inputWord.upper()} word is: \n{syns[0].definition()}")
print(f"Examples of {inputWord.upper()} word use: \n{syns[0].examples()}")
```
